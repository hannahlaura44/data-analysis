### Part 1: Data Analysis

#### A) Multiple Regression Analysis

I conducted a multiple regression analysis (ordinary least squares) to obtain an overview of which independent variables (season, holiday, weather, etc.) have a significant impact on the dependent variable (number of bikes rented).

I conducted multiple regression analyses including all the independent variables (except dteday) for three seperate dependent variables: total count, registered count, and casual count.
